```python
bert_embeddings = []
for i in range(data.shape[0]):
    try:
        inp = data['embeddings'][i].replace('\n','').replace(' dtype=float32','').replace(" ", "").strip()[7:-3]
        output_list = [float(x) for x in inp[1:-1].split(',')]
        assert len(output_list) == 768
        bert_embeddings.append(output_list)
    except:
        print(i)
```

```python
len(bert_embeddings)
```

```python
import numpy as np
from prettytable import PrettyTable

from sklearn.metrics.pairwise import cosine_similarity

# assume that your dataframe is called `df` and the embedding column is called `embedding`
bert_embeddings = np.stack(bert_embeddings)
# Create the table object
table = PrettyTable()
table.field_names = ["Category", "Title", "Similarity"]

# define a function to get the closest rows to a query embedding
def get_closest_rows(query_embedding, embeddings, titles,categories, k=10):
    cosine_similarities = cosine_similarity(query_embedding.reshape(1,-1), embeddings)
    closest_indices = cosine_similarities.argsort()[0][::-1][:k]
    closest_rows = [(categories[i], titles[i], cosine_similarities[0][i]) for i in closest_indices]
    return closest_rows

# assume that you have a query_embedding and want to find the closest rows in the dataframe
queries = list(bert_data[bert_data.category=='DS'].index)
q = 4911
query_embedding = bert_embeddings[q] # replace with your own query embedding
print("Input:")
print(f"query title - {bert_data['title'][q]}, \nInput Category - {bert_data['category'][q]}" )
closest_rows = get_closest_rows(query_embedding, bert_embeddings, bert_data['title'].values, bert_data['category'].values)

# Print the closest rows in a table format
# Add the data rows
print("Recommendation:")
count = 0
for cat, title, similarity in closest_rows[1:]:
    if cat == bert_data['category'][q]:
        count +=1
    table.add_row([cat, title, similarity])

# Print the table
print(table)
```
